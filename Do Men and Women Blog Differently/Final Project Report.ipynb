# Battle of the Sexes - Blogger Edition
***
*There is a stereotype is that women use more words than men. A common assertion is that women use 20,000 words a day while men use 7,000, yet there is no scientific basis for this claim. While subsequent studies have found negligible differences in the amount of words men and women speak per day, far less studies have been conducted on the gender differences in vernacular. In general, these studies have found that men speak in terms of the big picture, while women tend to focus on specifics. This was done by assigning scores of abstractness to specific words and measuring the frequency of these words in the speech and text of men and women. Besides just abstractness, I will seek some other form of NLP analyses that may highlight the differences between the words that males and females use. Current research has concluded that men speak in more abstract terms, while women are more focused on details. The results of my projects could be supplemental to the efforts of these researchers.*

## 1. Data
***
[Blog Authorship Corpus](https://u.cs.biu.ac.il/~koppel/BlogCorpus.htm)

The key data source came from the Blog Authorship Corpus, collected by Schler, Koppel, Argamon, and Pennebaker (2006). The Blog Authorship Corpus has been scraped from blogger.com, and includes 681,288 posts written by 19,320 bloggers. All blogs were gathered in August 2004, and ages of bloggers ranged from 13-47.

I conducted many explorations on the distributions of variables in the data, but one interesting analysis was finding a significant difference in sentiment between blogs written by females vs. males. Below is a visualization of these differences.



As we can see, female bloggers made up a majority of positive sentiment blogs, while male blogges made up a majority of negatively written blogs. With the help of a t-test, I was able to find that this difference was in fact significant. There were also significant differences found in sentiment analysis between the age groups of 13-17 and 22-28. However, there was no significant difference found in sentiment between any other age groups.



Lastly, a word cloud of blogs between different age groups showed no drastic differences in vocabulary used. This was concerning, as this was what I was going to try to use to predict whether blogs were written by men or women.



## 4. Algorithms and Machine Learning
***
[ML Notebook](https://github.com/saintgau/Springboard-Projects/blob/master/Springboard%20Projects/Battle%20of%20the%20Sexes%20-%20Blogger%20Edition/Pre-Processing%20and%20Modeling.ipynb)

I chose to work with Python's scikit-learn package for training my predictive model. Using a pipeline that included a TfidfVectorizer and StandardScaler, I tried both a Random Forest and an XGBoost Classifier. The XGBoost Classifier outperformed the Random Forest Classifier, so I ended up using the XGBoost Classifier.

I was not able to use the entire dataset, as my local machine was not able to handle it. As a result, I was only able to use blogs written by teens from the ages of 13-17.

## 5. Evaluations
***
The XGBoost Classifier came out to have an AUC of 0.744.
